```python
# prompt: I want to get correlations, heatmaps, causations from the loaded data

import seaborn as sns
import matplotlib.pyplot as plt

# Calculate the correlation matrix
corr_matrix = df.corr(numeric_only=True)

# Plot the heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

# Note: Causation cannot be determined from correlation alone.
# Correlation indicates a relationship between variables, but not
# whether one variable causes the other. More advanced techniques
# like causal inference methods (e.g., Granger causality, causal graphical models)
# would be needed to investigate potential causal relationships.
```

```python
# Get a list of numeric column names
numeric_columns = df.select_dtypes(include='number').columns.tolist()

# Specify the column to remove (replace 'column_to_remove' with the actual column name)
column_to_remove = 'CLIENTNUM' # Replace with the column you want to exclude

# Remove the specified column from the list if it exists
if column_to_remove in numeric_columns:
    numeric_columns.remove(column_to_remove)

# Iterate through the modified list of numeric columns and plot histograms
for i in numeric_columns:
  sns.histplot(data = df, x = i)
  plt.show()
```
